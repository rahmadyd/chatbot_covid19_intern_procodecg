```python
# Challenge: Bikin class sederhana yang nyimpen nama
class Person:
    def __init__(self, nama):
        self.nama = nama
        pass
    
    def sapa(self):
        print(f"Halo, saya {self.nama}")
        pass

# Test:
orang = Person("Rahmad")
orang.sapa()  # Output: "Halo, saya Rahmad"

orang =  Person("Saka")
orang.sapa()


```
