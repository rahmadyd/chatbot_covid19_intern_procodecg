```python
import PyPDF2
import pdfplumber

pdf_raw = "D:/chatbot_covid19_intern_procodecg/data/raw/knowladge_covid19_indonesia.pdf"

def identifikasi_pdf(pdf_raw):
    text_pdf = ""
    
    with open(pdf_raw, 'rb') as file:
        pdf = PyPDF2.PdfReader(file)
        
        for halaman in pdf.pages:
            teks = halaman.extract_text()
            teks_pdf += teks + "\n"
    
    return teks_pdf
```

```python
def potong_jadi_bab_kecil(teks_bersih):
    """
    Analogi: Memotong buku jadi bab-bab kecil (500-1000 kata)
    Overlap = seperti bab baru mulai sebelum bab lama selesai
    """
    
    # Parameter mudah
    UKURAN_CHUNK = 1000    # ±1000 karakter per potongan
    OVERLAP = 200          # 200 karakter tumpang tindih
    
    chunks = []
    start = 0
    
    while start < len(teks_bersih):
        # Ambil potongan
        end = start + UKURAN_CHUNK
        chunk = teks_bersih[start:end]
        
        # Simpan dengan metadata
        chunks.append({
            'text': chunk,
            'metadata': {
                'start_char': start,
                'end_char': end,
                'source': 'knowladge_covid19_indonesia.pdf'
            }
        })
        
        # Geser dengan overlap (seperti tumpang tindih)
        start += (UKURAN_CHUNK - OVERLAP)
    
    return chunks
```

```python
import json
def simpan_chunks_rapi(chunks, nama_file_output):
    """
    Simpan seperti map kecil-kecil yang rapi
    """
    with open(nama_file_output, 'w', encoding='utf-8') as f:
        for i, chunk in enumerate(chunks):
            # Tulis setiap chunk di baris terpisah (JSONL format)
            f.write(json.dumps(chunk, ensure_ascii=False) + '\n')
            
    print(f"✓ Disimpan {len(chunks)} chunks ke {nama_file_output}")
```
