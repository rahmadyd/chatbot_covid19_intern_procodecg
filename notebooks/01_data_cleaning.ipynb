```python
import pandas as pd
import json

# Baca data bersih
df = pd.read_csv("D:/chatbot_covid19_intern_procodecg/data/use/covid19_final_version.csv")

records = []
for _, row in df.iterrows():
    # Format persentase agar lebih rapi (dua angka di belakang koma)
    rasio_kematian = f"{row['Rasio Kematian Kasus']:.2f}%"
    rasio_sembuh = f"{row['Rasio Kesembuhan Kasus']:.2f}%"
    
    text = (
        f"Pada {row['Tanggal']}, di {row['Provinsi']}, tercatat {row['Kasus Baru']} kasus baru COVID-19, "
        f"{row['Kematian Baru']} kematian baru, dan {row['Sembuh Baru']} pasien sembuh. "
        f"Total kasus mencapai {row['Total Kasus']}, dengan {row['Total Kematian']} kematian dan "
        f"{row['Total Sembuh']} pasien sembuh. Rasio kematian kasus adalah {rasio_kematian}, "
        f"dan rasio kesembuhan adalah {rasio_sembuh}. "
        f"Jumlah penduduk wilayah ini adalah {row['Jumlah Penduduk']} orang."
    )
    
    records.append({"text": text})

# Simpan ke file JSONL
output_path = "D:/chatbot_covid19_intern_procodecg/data/processed/covid_19_to_narative.jsonl"
with open(output_path, "w", encoding="utf-8") as f:
    for r in records:
        f.write(json.dumps(r, ensure_ascii=False) + "\n")

print(f"‚úÖ File teks dengan format persentase disimpan di: {output_path}")
```
